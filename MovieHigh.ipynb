```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
import re
```

```python
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
stop_words=set(stopwords.words('english'))
lemmatizer=nltk.WordNetLemmatizer()
```

```python
def preprocess_text(text):
  text=str(text).lower()
  #remove pantuation
  text=re.sub(r'[^a-zA-Z\s]','',text)
  words=text.split()#word tokenisation
  words=[word for word in words if word not in stop_words]#remove stop words
  #lemmatize
  words=[lemmatizer.lemmatize(word) for word in words]
  return " ".join(words)
```

```python
df['tags']=df['tags'].apply(preprocess_text)
```

TFIDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf=TfidfVectorizer(max_features=50000,ngram_range=(1,2),stop_words='english')
```

```python
tfidf_metrix=tfidf.fit_transform(df['tags'])#all tag change in vetrix
```

```python
tfidf_metrix
```

COSINE SIMILARITYYYYY





```python
from sklearn.metrics.pairwise import cosine_similarity

def recommend(title,n=10):
  if title not in indices:
    return ["🎬movie not found❌"]

  idx=indices[title]
  sim_score=cosine_similarity(tfidf_metrix[idx],tfidf_metrix).flatten()
  similar_idx=sim_score.argsort()[::-1][1:n+1]
  return df['title'].iloc[similar_idx]
```

```python
recommend('Toy Story')
```
